```python

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
```

```python
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data.head()
```

```python
Year = crime_data.iloc[:,0]
Year
#def crimetime (title, column):

Violent_Rate = crime_data.iloc[:,2]
Murder_Rate = crime_data.iloc[:,5]
Agg_Ass_Rate = crime_data.iloc[:,11]
x_axis = Year
(slope, intercept, _, _, _) = stats.linregress(x_axis, Violent_Rate)
fit = slope * x_axis + intercept
#return [fit, slope, int]
```

```python
(slope, intercept, _, _, _) = stats.linregress(x_axis, Murder_Rate)
fit = slope * x_axis + intercept
```

```python
(slope, intercept, _, _, _) = stats.linregress(x_axis, Agg_Ass_Rate)
fit = slope * x_axis + intercept
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
fig.suptitle("Crime Rates Over Time", fontsize=16, fontweight="bold")
ax1.set_xlim(min(Year), max(Year))
ax1.plot(Year, Violent_Rate, linewidth=1, marker="o")
ax1.plot(Year, fit, "b--", linewidth=1)
ax1.set_ylabel("Violent Crime Rate")
```
